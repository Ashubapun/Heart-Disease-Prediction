```python
new_df = df.dropna()
```

```python
new_df = new_df.rename(columns = {'TenYearCHD' : 'Y'})
```

```python
new_df['Y'].value_counts()
```

```python
x = new_df.drop(columns = 'Y')
y = new_df['Y']
```

```python
# resampling
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(sampling_strategy='minority')
x,y = ros.fit_resample(x,y)
```

```python
x.shape
```

```python
y.shape
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 42)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
# modeling
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
score = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
score, cm
```

```python
# cross-validation
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(estimator = knn, X = x_train, y = y_train, cv = 10)
cvs.mean()
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=2000)          ## penalty = 'elasticnet', max_iter=1000, solver = 'saga'
lr.fit(x_train,y_train)
y_pred1 = lr.predict(x_test)

score1 = accuracy_score(y_test, y_pred1)
cm1 = confusion_matrix(y_test, y_pred1)

cvs1 = cross_val_score(estimator = lr, X = x_train, y = y_train, cv = 10)
score1, cm1, cvs1.mean()
```

```python
# svm classifier
from sklearn.svm import SVC
sv = SVC()
sv.fit(x_train,y_train)
y_pred2 = lr.predict(x_test)

score2 = accuracy_score(y_test, y_pred2)
cm2 = confusion_matrix(y_test, y_pred2)

cvs2 = cross_val_score(estimator = sv, X = x_train, y = y_train, cv = 10)
score2, cm2, cvs2.mean()
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion= 'gini')
dt.fit(x_train,y_train)
y_pred3 = lr.predict(x_test)

score3= accuracy_score(y_test, y_pred3)
cm3 = confusion_matrix(y_test, y_pred3)

cvs3 = cross_val_score(estimator = dt, X = x_train, y = y_train, cv = 10)
score3, cm3, cvs3.mean()
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=150)
rf.fit(x_train,y_train)
y_pred4 = lr.predict(x_test)

score4 = accuracy_score(y_test, y_pred4)
cm4 = confusion_matrix(y_test, y_pred4)

cvs4 = cross_val_score(estimator = rf, X = x_train, y = y_train, cv = 10)
score4, cm4, cvs4.mean()
```

```python
# Perceptron
from sklearn.linear_model import Perceptron
percept = Perceptron(alpha = 0.001)
percept.fit(x_train,y_train)
y_pred5 = lr.predict(x_test)

score5 = accuracy_score(y_test, y_pred5)
cm5 = confusion_matrix(y_test, y_pred5)

cvs5 = cross_val_score(estimator = percept, X = x_train, y = y_train, cv = 10)
score5, cm5, cvs5.mean()
```

```python
# XGBoost
from sklearn.ensemble import GradientBoostingClassifier
xgb = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
xgb.fit(x_train, y_train)
y_pred6 = xgb.predict(x_test)

score6 = accuracy_score(y_test, y_pred5)
cm6 = confusion_matrix(y_test, y_pred5)

cvs6 = cross_val_score(estimator = xgb, X = x_train, y = y_train, cv = 10)
score6, cm6, cvs6.mean()
```
